```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n=200
r =np.random.normal(0,1,100)
r2=np.random.normal(2,1,100)
r=np.concatenate((r,r2))
```

```python

class_num=3
e=np.random.randint(0,class_num,n)
```

```python
c=np.zeros([n,class_num])
for k,ek in enumerate(e):
    c[k,ek]=1

rlis=[]
for k in range(class_num):
    rlis.append(r[e==k])
    plt.scatter(rlis[k],np.zeros(rlis[k].shape)+k)
```

```python
mlis=[]
for k in range(class_num):
    mlis.append(np.mean(rlis[k]))




d=np.array([(r-m)**2 for m in mlis])

e=np.argmin(d,axis=0)
```

```python
m0=[0,1]
m1=[3,2]
s0=[[2,1],[2,2]]
s1=[[2,-3],[1,2]]
r =np.random.multivariate_normal(m0,s0,1000)
r2=np.random.multivariate_normal(m1,s1,1000)
r=np.concatenate((r,r2))

```

```python
n=len(r)
```

```python
plt.scatter(r[:,0],r[:,1])
```

```python
rlis=[]
for k in range(class_num):
    rlis.append(r[e==k,:])
    plt.scatter(rlis[k][:,0],rlis[k][:,1])
```

```python
from PIL import Image
```

```python
img=np.array(Image.open('Lenna.png'))
```

```python
r=img.reshape(512*512,3)
```

```python
n=len(r)
```

```python
class_num=2
e=np.random.randint(0,class_num,n)
loss=[]
loss_img=[]
mlis=np.zeros([class_num,3])
```

```python
rlis=[]
for k in range(class_num):
    rlis.append(r[e==k,:])
    
for k in range(class_num):
    if len(rlis[k])==0:
        continue
    mlis[k,:]=np.mean(rlis[k],axis=0)

d=np.array([np.sum((r-m)**2,axis=1) for m in mlis])
e=np.argmin(d,axis=0)
loss.append(np.sum(np.min(d,axis=0)))

img_r=np.zeros(r.shape)
img_r=np.array([mlis[ek] for ek in e])
img_r=img_r.reshape(512,512,3)
img_r=np.uint8(img_r)

plt.subplot(1,3,1)
plt.imshow(img_r)
plt.subplot(1,3,2)
plt.plot(loss)

loss_img.append(np.sum(np.abs(img_r-img)))

plt.subplot(1,3,3)
plt.plot(loss_img)
```
